```python
from collections import Counter
import nltk
import string
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

text = input()

# step 1: lowercase
text = text.lower()

# step 2: remove punctuation
for t in string.punctuation:
    text = text.replace(t, "")

# step 3: tokenization
words = word_tokenize(text)
print("Tokens:\n", words)

# step 4: total words
total_words = len(words)
print("Total Words:\n", total_words)

# step 5: vocabulary size
unique_words = set(words)
print("Vocabulary Size:", len(unique_words))

# step 6: word frequency
freq = Counter(words)
print("\nWord Frequencies:\n", freq)

# step 7: top 10 words
print("\nTop 10 words")
for word, count in freq.most_common(10):
    print(word, ":", count)


hapax_legomena = []
for word,count in freq.items():
    if count == 1:
        hapax_legomena.append(word)
print("\nRare words=",hapax_legomena)


# step 8: TTR
ttr = len(unique_words) / total_words
print("\nType Token Ratio =", round(ttr, 2))

# step 9: stopword removal
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w not in stop_words]
print("\nFiltered Words =", filtered_words)

# step 10: frequency after stopword removal
filtered_freq = Counter(filtered_words)
print("Filtered words frequency count =", filtered_freq)

# step 11: chart
word_list = list(filtered_freq.keys())
counts_list = list(filtered_freq.values())

plt.figure(figsize=(8, 5))
plt.bar(word_list, counts_list)
plt.title("NLP word frequency analysis")
plt.xlabel("words")
plt.ylabel("frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
from collections import defaultdict,Counter

#training corpus
text ="""
What is Python
What is Libraries
What is LLM
What is LLM
what is Pandas 
What is Data
"""

#tokenization
words = text.lower().split()

#build bigram
model = defaultdict(list)
for i in range(len(words)-2):
	w1,w2,w3 = words[i],words[i+1],words[i+2]
	model[(w1,w2)].append(w3)

#Input Phrase
input_words = ("what","is")

#predict
if input_words in model:
    prediction = Counter(model[input_words]).most_common(1)[0][0]
    print("Input:What is")
    print("Next word prediction:",prediction)
else:
    print("No prediction found")
```
