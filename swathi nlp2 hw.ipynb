```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk
from nltk.tree import Tree
from nltk.classify import NaiveBayesClassifier
from nltk.probability import FreqDist
from nltk.util import bigrams, trigrams
from nltk.corpus import wordnet
from nltk.metrics import precision
```

```python
train_data = [
    ({'ai': True, 'technology': True}, 'Technology'),
    ({'cricket': True, 'match': True}, 'Sports'),
    ({'business': True, 'market': True}, 'Business'),
    ({'government': True, 'election': True}, 'Politics')
]
```

```python
classifier = NaiveBayesClassifier.train(train_data)
```

```python
test = {
    'ai': 'ai' in news.lower(),
    'technology': 'technology' in news.lower(),
    'cricket': 'cricket' in news.lower(),
    'business': 'business' in news.lower(),
    'government': 'government' in news.lower(),
    'election': 'election' in news.lower()
}
```

```python
category = classifier.classify(test)
print("Category :", category)
```

```python
test_set = [
    ({'ai': True, 'technology': True}, 'Technology'),
    ({'business': True, 'market': True}, 'Business')
]

accuracy = nltk.classify.accuracy(classifier, test_set)

reference = {"Technology", "Business"}
predicted = {"Technology", "Business"}

pre = precision(reference, predicted)

print("Accuracy :", round(accuracy * 100, 2), "%")
print("Precision :", round(pre * 100, 2), "%")
```

```python
words = [w.lower() for w in tokens if w.isalpha()]
fd = FreqDist(words)

for word, count in fd.items():
    print(word, ":", count)
```

```python
total = len(words)

for word in ["ai", "technology", "business"]:
    print(f"P({word}) = {fd[word]}/{total} = {fd[word]/total:.2f}")
```

```python
for word in ["technology", "business", "platform"]:
    syn = wordnet.synsets(word)
    if syn:
        print(word, "->", syn[0].definition())
```

```python
for bg in bigrams(words):
    print(bg)
```

```python
for tg in trigrams(words):
    print(tg)
```
